```python
import pandas as pd

df = pd.read_csv("./datas_csv/data.csv")

df.head()
```

```python
dinosaurios_argentinos = df[df["lived_in"] == "Argentina"]
dinosaurios_argentinos.head(3)
```

```python
dinosaurios_argentinos["especie_nombre"] = dinosaurios_argentinos["name"]+ "-" +dinosaurios_argentinos["species"]
dinosaurios_argentinos.head(3)
```

```python
especie_dinos_argentinos = dinosaurios_argentinos["especie_nombre"]
especie_dinos_argentinos.head()
```

```python
longitud_dinos_argentinos = dinosaurios_argentinos["length"].str.rstrip('m').astype(float)

longitud_dinos_argentinos.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.title("Mayor longitud de los dinosaurios que habitaron Argentina", fontsize=10, fontweight= "bold")
plt.bar(especie_dinos_argentinos, longitud_dinos_argentinos, color="white", edgecolor="red")
plt.ylabel("Longitud")
plt.xlabel("Nombre del dinosaurio", fontsize=7)
plt.ylim(0,38)
plt.xticks(rotation=70, ha='right')
for i, valor in enumerate(longitud_dinos_argentinos):
    plt.text(i, valor + 0.5, f'{valor}', ha="center")
plt.show()
```

```python
dinosaurios_carnivoros = dinosaurios_argentinos[dinosaurios_argentinos["diet"] == "carnivorous"]
cantidad_dinos_carnivoros = len(dinosaurios_carnivoros["diet"])

dinosaurios_hervivoros = dinosaurios_argentinos[dinosaurios_argentinos["diet"] == "herbivorous"]
cantidad_dinos_hervivoros = len(dinosaurios_hervivoros["diet"])

dinosaurios_omnivoros = dinosaurios_argentinos[dinosaurios_argentinos["diet"] == "omnivorous"]
cantidad_dinos_omnivoros = len(dinosaurios_omnivoros["diet"])

dinosaurios_dietas = pd.DataFrame([[cantidad_dinos_carnivoros], [cantidad_dinos_hervivoros],[cantidad_dinos_omnivoros]])
dinosaurios_dietas["Tipo_alimentacin"] = ["Carnivoro", "Hervivoros","Omnivoros"]
dinosaurios_dietas = dinosaurios_dietas.rename(columns={0: "cantidad_en_Argentina"})
dinosaurios_dietas
```

```python
import matplotlib.pyplot as plt

cantidad_en_Argentina = dinosaurios_dietas["cantidad_en_Argentina"]
tipo_de_alimentacion = dinosaurios_dietas["Tipo_alimentacin"]

plt.figure(figsize=(12, 6))
plt.pie(cantidad_en_Argentina, labels=tipo_de_alimentacion, autopct="%1.1f%%", startangle=140)
plt.title("Distribucin de dinosaurios por tipo de alimentacin en Argentina")
plt.show()
```

```python
top_paises_especies = df.groupby('lived_in')['name'].nunique().reset_index()

top_paises_especies = top_paises_especies.rename(columns={'name': 'cantidad_de_especies'})

top_10_paises_especies = top_paises_especies.sort_values(by='cantidad_de_especies', ascending=False).head(10)

top_10_paises_especies
```

```python
import matplotlib.pyplot as plt

yCantidad_de_especies = top_10_paises_especies["cantidad_de_especies"]
xPaises = top_10_paises_especies["lived_in"]

plt.figure(figsize=(12, 8))
plt.bar(xPaises, yCantidad_de_especies, color='#6BAED6', edgecolor='black')


plt.title("Top 10 pases con ms especies de dinosaurios", fontsize=18, fontweight='bold')
plt.xlabel("Cantidad de Especies", fontsize=14)
plt.ylabel("Pases", fontsize=14)
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12)
for i, valor in enumerate(yCantidad_de_especies):
    plt.text(i, valor + 0.5, f'Top:{i+1}: {valor}', ha="center", fontsize=10)


plt.show()
```
